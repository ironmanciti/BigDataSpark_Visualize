```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## DataFrame.plot

```python
# DataFrame 생성
# - A, B, C라는 세 개의 열을 포함하는 데이터프레임을 생성합니다.
# - 각 열은 365개의 난수로 이루어진 누적합(cumsum) 데이터로 생성됩니다.
# - B는 A에 20을 더한 데이터, C는 A에서 20을 뺀 데이터로 구성됩니다.
# - 데이터의 인덱스는 2021년 1월 1일부터 시작하여 365일 동안의 날짜로 설정됩니다.

df = pd.DataFrame({"A": np.random.randn(365).cumsum(),  # A 열: 난수 누적합
                    "B": np.random.randn(365).cumsum() + 20,  # B 열: 난수 누적합 + 20
                    "C": np.random.randn(365).cumsum() - 20},  # C 열: 난수 누적합 - 20
                    index=pd.date_range('1/1/2021', periods=365))  # 날짜 인덱스 생성
df.head()
```

```python
# 모든 column 을 동시에 plot
df.plot()
```

```python
# 특정 column 을 x 축, y 축으로 지정
df.plot('A', 'B', kind='scatter');
```

- 2 차원 평면상의 3 차원 정보 plot

```python
# Pandas 내장 시각화 기능 사용
# - 'A' 열과 'B' 열을 x축과 y축으로 설정하여 산점도(scatter plot)를 생성합니다.
# - 점의 색상은 'C' 열의 값에 따라 설정되며, 색상 맵은 'viridis'를 사용합니다.
df.plot.scatter('A', 'B', c='C', colormap='viridis');
```

```python
# matplotlib 사용
# - 산점도(scatter plot)를 생성합니다.
# - x축 데이터는 df.A (DataFrame의 A 열)입니다.
# - y축 데이터는 df.B (DataFrame의 B 열)입니다.
# - 점의 색상은 df.C (DataFrame의 C 열) 값을 기반으로 합니다.
# - 색상 맵은 'viridis'를 사용합니다.
plt.scatter(df.A, df.B, c=df.C, cmap='viridis')

# 색상 막대(colorbar)를 추가하여 C 열 값의 색상 범위를 표시합니다.
plt.colorbar()
```

```python
# box plot
df.plot.box();
```

```python
# histogram
df.plot.hist(alpha=0.7);
```

가우스 커널을 사용하여 커널 밀도 추정 플롯 생성

```python
df.plot.kde();
```

## The Dataset: Immigration to Canada from 1980 to 2013 <a id="2"></a>

Dataset Source: [International migration flows to and from selected countries - The 2015 revision](http://www.un.org/en/development/desa/population/migration/data/empirical2/migrationflows.shtml).

이 데이터 세트에는 목적지 국가에서 기록한 국제 이민자의 흐름에 대한 연간 데이터가 포함되어 있습니다. 자료는 출생지, 시민권 또는 외국인과 국민 모두를위한 이전 / 다음 거주지에 따라 유입과 유출을 모두 나타냅니다. 현재 버전은 45 개국에 관한 데이터를 제공합니다.

이 실습에서는 캐나다 이민 데이터에 중점을 둘 것입니다.

```python
import pandas as pd
import matplotlib.patches as mpatches  # 와플 차트를 그리기 위해 필요

# 캐나다로의 이민 데이터를 담고 있는 Excel 파일을 읽어 DataFrame으로 변환
df_can = pd.read_excel(
    'BigDataDatasets/Canada.xlsx',  # 파일 경로
    engine='openpyxl',  # openpyxl 엔진을 사용하여 Excel 파일 읽기
    sheet_name='Canada by Citizenship',  # 읽을 시트 이름
    skiprows=range(20),  # 상단 20행 생략
    skipfooter=2  # 하단 2행 생략
)

df_can.head(3)
```

```python
df_can.columns
```

```python
# 데이터 세트를 정리하여 불필요한 열(예 : REG)을 제거
df_can.drop(['Type', 'Coverage', 'AREA', 'REG', 'DEV'], 
            axis=1, inplace=True)
df_can.head(2)
```

```python
# 이해하기 쉽게 열 이름을 변경합니다.
df_can.rename(columns={'OdName': 'Country', 
                       'AreaName': 'Continent', 
                       'RegName': 'Region'},
              inplace=True)
df_can.head(2)
```

```python
# 일관성을 위해 모든 열 레이블을 문자열 유형으로 만듭니다.
df_can.columns = [str(col) for col in df_can.columns]
df_can.columns
```

```python
# 국가 이름을 색인으로 설정하면 .loc 메소드를 사용하여 국가를 빠르게 찾는 데 유용합니다.
df_can.set_index('Country', inplace=True)

# 이 노트북에서 사용할 연도 - 나중에 플로팅하는 데 유용합니다.
years = [str(y) for y in range(1980, 2014)]
print(years)
```

```python
# total column 추가
df_can['Total'] = df_can.iloc[:, 3:].sum(axis=1)
df_can.head()
```

```python
df_can.shape
```

## Series or DataFrame 의 plotting

- X 축 : index value   

- Y 축 : 해당 index 의 value

- plot 의 kind

    * `line` (default)
    * `bar` : vertical bar plots
    * `barh` :  horizontal bar plots
    * `hist` : histogram
    * `box` : boxplot
    * `kde` or `density` : density plots
    * `area` : area plots
    * `pie` : pie plots
    * `scatter` : scatter plots
    * `hexbin` : hexbin plot

### Line Pots (Series/Dataframe) 

- Haiti 의 연도별 이민자 수 시각화

```python
# 연도별 이민자수 column만 선택
haiti = df_can.loc['Haiti', years] 
haiti.head()
```

```python
plt.figure(figsize=(8, 4))

# 플로팅을 위해 인덱스 값을 정수 유형으로 변경
# 기존 인덱스가 문자열 또는 다른 유형인 경우 이를 정수형으로 변환
haiti.index = haiti.index.map(int)

# 라인 플롯 생성
haiti.plot(kind='line')

# 현재 플롯의 축(Axis) 객체를 가져옴
ax = plt.gca()

ax.set_xlabel('Years')  
ax.set_ylabel('이민자 수')  

# 플롯 제목 설정
ax.set_title('Haiti로부터의 이민자 수')  

# 특정 사건에 대한 텍스트 표시
# x 좌표, y 좌표 위치에 텍스트 삽입
ax.text(2010, 6000, '2010년 지진')
plt.show()
```

### 다른 나라들 간의 이민자 수 비교를 위해 그래프에 더 많은 국가 추가 

- 1980 년부터 2013 년까지 인도와 중국 이민자 수를 비교

Step 1: China 와 India 를 dataframe 에서 따로 추출하여 plot 한다

```python
# 'China'와 'India' 두 나라의 이민자 데이터를 선택하여 새로운 데이터프레임 생성
# 'years' 변수는 열 이름 리스트로, 1980년부터 2013년까지의 연도를 포함
df_ci = df_can.loc[['China', 'India'], years]
df_ci
```

* *pandas*는 x 축의 인덱스와 y 축의 개별 라인으로의 열을 그려준다. `df_ci`는`country`를 인덱스로,`years`를 컬럼으로하는 데이터 프레임이기 때문에 행과 열을 바꾸기 위해 먼저`transpose ()`메소드를 사용하여 데이터 프레임을 변경한다.

```python
# 데이터프레임 전치(행과 열을 뒤바꿈)
df_ci = df_ci.transpose()

# 플로팅을 위해 df_ci의 인덱스 값을 정수형으로 변환
df_ci.index = df_ci.index.map(int)

df_ci.head()
```

```python
# 라인 플롯(Line Plot) 생성
df_ci.plot(kind='line')  

# x축의 눈금(xticks)을 5년 단위로 설정
plt.xticks(df_ci.index[::5]) 

# x축 레이블 설정
plt.xlabel('Years') 

plt.show()
```

### 캐나다 이민에 가장 많은 기여를 한 상위 5 개국의 추세를 비교

```python
# Pandas 옵션 설정: 모든 행을 출력하도록 설정 (주석 처리됨)
# pd.set_option('display.max_row', None)

# 'years' 열에 해당하는 데이터프레임의 모든 행에 대해 합계를 계산
# 각 국가별로 연도 데이터를 합산하여 총 이민자 수를 계산
top5 = df_can.loc[:, years].sum(axis=1)  

# 이민자 수가 많은 순서대로 내림차순 정렬
top5 = top5.sort_values(ascending=False)  

top5
```

```python
# 특정 국가 이름 변경
df_can.rename(index={'United Kingdom of Great Britain and Northern Ireland': 'Ireland'}, inplace=True)

# 변경 확인
df_can.loc['Ireland']
```

```python
# 상위 5개 국가의 이민자 데이터를 선택하여 데이터프레임 생성
df_top5 = df_can.loc[
    ['India', 'China',  # 인도, 중국
     'Ireland',  # 아일랜드
     'Philippines',  # 필리핀
     'Pakistan'],  # 파키스탄
    years  # 선택한 연도(years) 데이터만 포함
]

# 데이터프레임 전치(transpose) 수행
df_top5 = df_top5.transpose()  # 행과 열을 뒤바꿔 연도를 행 인덱스로 변경

df_top5.head()
```

```python
df_top5.index
```

- 상위 5 개의 인덱스 값을 정수형으로 변경하여 플로팅

```python
# df_top5 데이터프레임의 인덱스를 정수형(int)으로 변환
df_top5.index = df_top5.index.map(int) 

df_top5.head()
```

```python
# df_top5 데이터를 선형 그래프로 시각화
df_top5.plot(kind='line')  

# 현재의 Axes 객체 가져오기
ax = plt.gca()

# 범례(legend) 설정: 국가 이름을 표시
ax.legend(['India', 'China', 'UK', 'Philippines', 'Pakistan'])

# x축의 눈금(tick) 설정: 5년 단위로 눈금 표시
ax.set_xticks(df_top5.index[::5])

# x축 레이블 설정
ax.set_xlabel('Years')

# y축 레이블 설정
ax.set_ylabel('Number of Immigrants')
```

## Bar Chart + 주석 달기

- 그룹화된 범주형 변수의 크기를 표시

- kind='bar' : 세로막대  
- kind='barh' : 가로막대

```python
df_iceland.plot(kind='bar', figsize=(8, 6))

plt.title('1980년부터 2013년까지 아이슬란드 이민자 수')
plt.ylabel('이민자 수')
plt.xlabel('연도')

#화살표 그리기
plt.annotate('',                      # 표시할 text
                   xy=(32, 70),       #  화살표의 머리 위치
                   xytext=(28, 20),   # 화살표의 꼬리 위치
                   arrowprops=dict(arrowstyle='->', color='red', lw=2)
            )
#글씨 쓰기
plt.annotate('2008 - 2012 세계 금융위기',      # 표시할 text
                  xy=(28, 30),                   # text 시작 위치
                  rotation=77,                   # trial and error 로 맞춤
                  va='bottom',                   # text 가 vertically bottom 정렬
                  ha='left'                      # text 가 horizontally 좌측 정렬
            )
plt.show()
```
