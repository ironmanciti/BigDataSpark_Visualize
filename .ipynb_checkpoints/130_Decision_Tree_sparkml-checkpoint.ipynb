# 030.Decision Tree

### iris dataset

iris.feature_names : 

Sepal Length : 꽃받침 길이  
Sepal Width  : 꽃받침 너비   
Petal Length  : 꽃잎 길이  
Petal Width   :  꽃잎 너비 

Species (꽃의 종류) :  setosa / versicolor / virginica 의 3종류로 구분된다.

**위 feature 를 모두 가지고 Decision Tree 알고리즘을 이용하여 꽃의 종류 분류**

```python
import findspark
findspark.init()

from pyspark.sql import SparkSession

spark = SparkSession.builder.appName("DataFrame").config("spark.sql.repl.eagerEval.enabled", True).getOrCreate()
spark
```

```python
from sklearn.datasets import load_iris
import pandas as pd
from pyspark.ml.classification import DecisionTreeClassifier
import warnings
warnings.filterwarnings('ignore')

iris = load_iris()
df_iris = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df_iris.head()
```

```python
df_iris['Species'] = iris.target
df_iris.head()
```

```python
df_iris = spark.createDataFrame(df_iris)
df_iris.show(5)
```

```python
from pyspark.sql.functions import col

df = df_iris.select(col("sepal length (cm)").alias("sepal_length"), col("sepal width (cm)").alias("sepal_width"), 
                 col("petal length (cm)").alias("potal_length"), col("petal width (cm)").alias("petal_width"), "Species")
df.show(5)
```

```python
df.printSchema()
```

```python
df.columns
```

```python
from pyspark.ml.feature import VectorAssembler

assembler = VectorAssembler(inputCols=df.columns[:-1], outputCol="features")
data = assembler.transform(df)
data.show(5)
```

### train / test split & accuracy check

```python
train_data, test_data = data.randomSplit([0.8, 0.2])
train_data.count(), test_data.count()
```

```python
dt = DecisionTreeClassifier(labelCol="Species", maxDepth=5).fit(train_data)
dt
```

```python
print(dt.featureImportances)
```

```python
## Prediction 
pred = dt.transform(test_data)
pred.show()
```

```python
from pyspark.ml.evaluation import MulticlassClassificationEvaluator

dt_auc = MulticlassClassificationEvaluator(labelCol="Species").evaluate(pred)
dt_auc
```

### Decision Tree 시각화

```python
from sklearn.datasets import load_iris
from sklearn import tree

iris = load_iris()
iris.data.shape, iris.target.shape
```

```python
clf = tree.DecisionTreeClassifier(max_depth=2)
clf.fit(iris.data, iris.target)
```

```python
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(25,20))
_ = tree.plot_tree(clf, 
                   feature_names=iris.feature_names,  
                   class_names=iris.target_names,
                   filled=True)
```
