# 탄자니아의 물 펌프 데이터 분석

1. **결측값 처리**:
   - 데이터를 정제하여 결측값을 효율적으로 처리하고, 분석 가능한 상태로 만듭니다.
   - Null 값이 많은 열을 식별하고 제거하거나 대체합니다.

2. **데이터 변환 및 재분류**:

3. **그룹화 및 집계**:

## 데이터 정제 (Data Cleaning)
데이터셋의 품질을 향상시키고 분석 및 모델링에 적합한 상태로 만들기 위해 수행

-------------
데이터 column 중 `region_code`와 `district_code` 는 숫자형 처럼 보이지만 실제로는 카테고리 데이터로 사용됩니다. 숫자형으로 유지하면 불필요하게 수학적 연산이 일어날 수 있으므로 문자열(StringType)로 변환하여 분석 및 시각화에 적합하게 타입을 변형 시킵니다.

```python
# 데이터프레임 전체 스키마 중 특정 열만 필터링하여 출력
schema_info = [field for field in data.schema.fields if field.name in ["region_code", "district_code"]]
for field in schema_info:
    print(f"Column: {field.name}, Type: {field.dataType}")
```

```python
from pyspark.sql.functions import col         # 특정 열을 참조하거나 조작할 때 사용하는 함수
from pyspark.sql.types import StringType  # 열의 데이터 타입을 문자열(String)로 지정하기 위해 사용하는 클래스

# 'region_code' 컬럼의 데이터 타입을 문자열로 변환
data = data.withColumn("region_code", col("region_code").cast(StringType()))

# 'district_code' 컬럼의 데이터 타입을 문자열로 변환
data = data.withColumn("district_code", col("district_code").cast(StringType()))

schema_info = [field for field in data.schema.fields if field.name in ["region_code", "district_code"]]
for field in schema_info:
    print(f"Column: {field.name}, Type: {field.dataType}")

# 데이터의 상위 5개 행 출력
print(data.show(5))
```

```python
# 중복된 id의 행 제거
data = data.dropDuplicates(["id"])

# 데이터의 전체 행 수 출력
data.count()
```

### 문자열 컬럼에서 공백을 제거
- 데이터의 정확성 확보 - 문자열 데이터에서 불필요한 공백이 포함되어 있으면, 동일한 값이더라도 서로 다른 값으로 인식.   
예) "ABC"와 " ABC "는 다른 값으로 처리
- 데이터 분석 및 비교 작업에서의 오류 방지 - 그룹화, 조인, 피벗 작업 시 예상치 못한 오류 발생

```python
# 데이터프레임의 모든 문자열 타입 컬럼명을 추출
str_cols = [item[0] for item in data.dtypes if item[1].startswith("string")]
print(str_cols)

from pyspark.sql.functions import trim

# 각 문자열 컬럼에 대해 양쪽 공백을 제거
for col_name in str_cols:
    data = data.withColumn(col_name, trim(data[col_name]))
```

### 결측치 제거

다음 스크립트는 PySpark 데이터프레임의 각 열에 대해 결측치 비율(Null 및 NaN 값의 비율)을 계산하는 데 사용됩니다. 결과적으로, 각 열의 결측치 비율을 딕셔너리 형태로 저장합니다. 이를 통해 데이터 정제 및 전처리 과정에서 결측치 처리의 우선순위를 결정할 수 있습니다.


```python
from pyspark.sql.functions import isnan, col, count, when

# 숫자형 데이터 타입
numeric_types = ["int", "bigint", "float", "double", "decimal"]

# 데이터 타입에 따라 isnan() 또는 isNull() 적용
agg_row = data.select([
    (
        # 숫자형 데이터 타입이면 isnan() + isNull(), 그 외 데이터 타입이면 isNull()만 사용
        count(when(
            (col(c).isNull() | isnan(c)) if data.schema[c].dataType.typeName() in numeric_types else col(c).isNull(),
            c
        )) / data.count()
    ).alias(c)
    for c in data.columns
]).collect()

# 결과 출력
agg_dict_list = [row.asDict() for row in agg_row]
agg_dict = agg_dict_list[0]
agg_dict
```

```python
# 결측치 비율이 높은 순으로 정렬
sorted_columns = sorted(agg_dict.items(), key=lambda x: x[1], reverse=True)

print("결측치 비율이 높은 순으로 정렬된 컬럼:")
for column, null_ratio in sorted_columns:
    print(f"Column: {column}, Null Ratio: {null_ratio:.2%}")
```

47% 의 결측치가 있는 `scheme_name` column 은 삭제

```python
# 모든 열에 대해 결측치가 있는 행 제거
data_cleaned = data.drop("scheme_name")
```

### `water_quality`의 분포 파악

```python
data.select('water_quality').distinct().show()
```

```python
# 'water_quality' 열을 기준으로 데이터프레임을 그룹화하여 각 그룹의 개수를 계산하고, 이를 내림차순으로 정렬하여 출력
data.groupBy("water_quality").count().orderBy("count", ascending=False).toPandas()
```

### status_group (펌프의 작동 상태)를 별로 지역 (region) 별 시간당 물 공급량 (amount_tsh) 파악

```python
# 'status_group' 열을 기준으로 데이터 그룹화하고,
# 'region' 열의 고유 값을 열로 변환하여 피벗 테이블을 생성한 뒤,
# 'amount_tsh' 값의 합계를 계산하여 출력
data.groupBy("status_group").pivot("region").sum("amount_tsh").toPandas()
```

```python
# 시각화를 위한 상태별 색상 정의
# 각 물 펌프의 상태(status_group)에 따라 색상을 지정
color_status = {
    "functional": "green",                # 정상 작동하는 펌프는 녹색으로 표시
    "non functional": "red",              # 작동하지 않는 펌프는 빨간색으로 표시
    "functional needs repair": "blue"     # 작동은 하지만 수리가 필요한 펌프는 파란색으로 표시
}
```

```python
# 시각화를 위한 특정 컬럼 선택 및 Pandas 데이터프레임으로 변환

# 1. 필요한 컬럼 정의
cols = ["status_group", "payment_type", "longitude", "latitude", "gps_height"]

# 2. 정의한 컬럼만 선택하여 PySpark 데이터프레임에서 추출 후 Pandas 데이터프레임으로 변환
df = data.select(cols).toPandas()
df.head()
```

### `payment_type (결제 방식)` 에 따른  `status_group (펌프 상태)`의 분포를 시각화

특정 결제 방식이 펌프 상태에 어떤 영향을 미치는지 패턴을 파악

```python
import matplotlib.pyplot as plt
import seaborn as sns

fig, ax = plt.subplots(figsize=(8, 6))

# - x축: 'payment_type' (결제 방식)
# - 막대 색상 구분: 'status_group' (펌프 상태)
sns.countplot(
    x="payment_type", hue="status_group", data=df, ax=ax, palette=color_status
)
plt.xticks(rotation=45)
plt.show()
```

#### 펌프 상태별 지리적 분포 시각화

- 물 펌프의 상태별로 색상을 구분하여 시각화
